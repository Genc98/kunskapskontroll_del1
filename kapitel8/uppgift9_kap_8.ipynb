```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

9. I denna uppgift utgår vi ifrån kodexempel 2 i detta kapitel. 

a) Ta egna bilder som du predikterar med en förtränad modell. 

Använder en förtränad modell på imagenet som heter ResNet50.

```python
model = ResNet50(weights='imagenet')
```

Deklarerar image paths.

```python
img_paths = ['bilder/kopp.jpg', 'bilder/vas.jpg']
```

Läser in bilderna, skalar, konverterar de till arrayer, förbehandlar de för ResNet50 och sparar dem i en lista som används sedan i modellen.

```python
imgs = []
for path in img_paths:
    img = image.load_img(path, target_size=(224,224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    imgs.append(x)
```

Samlar allt i en batch så att modellen predikterar allt samtidigt.

```python
batch = np.vstack(imgs)
```

Predikterar vad som finns bilderna.

```python
preds = model.predict(batch, verbose=2)
```

Sannolikheten visar vad som finns i bilderna. Den första bilden innehåller ett enkelt objekt som är lätt att identifiera, vilket ger högre sannolikhet, medan den andra bilden är svårare att tolka och därför ger ett lägre resultat.

```python
for i, pred in enumerate(preds):
    print(f'Predicted for image {i+1}:', decode_predictions(pred.reshape(1,1000), top=3)[0])
```
